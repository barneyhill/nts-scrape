```python
## add urls to the json:

import json
import tqdm

# Step 1: Load the JSON file
with open('../data/nts_17_7_23_tracklists.json') as json_file:
    data = json.load(json_file)

# Step 2: Read `links.txt` file
links_dict = {}
with open('../data/links.txt') as links_file:
    for line in links_file:
        links_dict[line.split("/")[-1].strip()] = line.strip()

# Step 3: Iterate over JSON file and add "url" key
for key in tqdm.tqdm(data.keys()):
    data[key]['url'] = links_dict.get(key)  # Add "url" key to the record

# Save the updated data back to the JSON file
with open('../data/nts_17_7_23_tracklists_with_urls.json', 'w') as json_file:
    json.dump(data, json_file, indent=4)
```
